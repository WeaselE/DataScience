```python
# Source: https://catalog.data.gov/dataset/conditions-contributing-to-deaths-involving-coronavirus-disease-2019-covid-19-by-age-group

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Conditions_Contributing_to_COVID-19_Deaths__by_State_and_Age.csv')
df.describe()
```

```python
df.head()
```

```python
df['Date'] = df['End Date']
df.drop(columns=['Month', 'Year', 'State', 'Flag', 'Data As Of', 'Start Date', 'End Date'], axis=1, inplace=True)
df = df[['Date', 'Condition Group', 'Condition', 'Age Group', 'COVID-19 Deaths', 'Number of Mentions']]
```

```python
print(df.head())
test = df.sample(n=200)
print(df.dtypes)
```

```python
test = test.groupby('Age Group')
test.head()
```
